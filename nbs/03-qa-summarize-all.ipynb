```python
import pandas as pd

df = pd.read_json('isrpa-api-full.json')
summary = list(df['summary'])
```

```python
summary[1]
```

```python
with open('isrpa-api-full.txt', 'w') as f:
    for item in summary:
        if item == '': continue
        text = f'name: {item["name"]}  parameters: '
        for p in item['parameters']:
            if isinstance(p, str):
                text += f'{p}'
            else:
                text += f'{p["name"]}'
            if 'description' in p:
                text += f'({p["description"]})'
            text += ' '
        text += f'\ndescription: {item["description"]} examples:'
        if isinstance(item['usages'], str):
            text += f'{item["usages"]} \n'
        else:
            for w in item['usages']:
                text += f'{w} \n'
        print(text)
        text += '\n'
        f.write(text)
```
