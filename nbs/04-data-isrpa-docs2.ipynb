```python
import pandas as pd

df = pd.read_json('isrpa-api.json')
articles = df['url'].str.contains('article')
df.loc[articles].head(10)
```

```python

import time
import random
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By


options = Options()
options.add_argument('--headless')


def scroll_by_func(driver):
    # get the initial scroll height of the page
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # wait for new content to load
        time.sleep(random.random() * 5)

        # get the new scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        #print(new_height)

        # check if the scroll height has changed
        if new_height == last_height:
            # if the scroll height hasn't changed, all content is loaded
            break

        # update the scroll height
        last_height = new_height

def get_article(new_url):
    driver = webdriver.Chrome(service=Service("/usr/bin/chromedriver"), options=options)
    driver.get(new_url)
    time.sleep(30)
    scroll_by_func(driver)
    #driver.save_screenshot(f'{article[0]}.png')
    time.sleep(random.random() * 3)
    try:
        content = driver.find_element(By.ID, 'editor-md').text
    except:
        content = driver.find_element(By.CLASS_NAME, 'article-content').text
    print(new_url, content)
    # driver.save_screenshot('isrpa-api-doc.png')
    # article-content
    driver.quit()
    return content
```

```python
options = Options()
options.add_argument('--headless')

driver = webdriver.Chrome(service=Service("/usr/bin/chromedriver"), options=options)
```

```python
url = df.loc[articles, 'url'].iloc[0]
print(url)

# navigate to a webpage
driver.get(url) # 微信公众平台网址
scroll_by_func(driver)

# help(driver)
driver.save_screenshot('video-tutorial.png')

content = driver.find_element(By.CSS_SELECTOR, 'blockquote p a')
print(content.text, content.get_property('href'))
```

```python
def get_real_url(url):
    driver = webdriver.Chrome(service=Service("/usr/bin/chromedriver"), options=options)
    driver.get(url)
    time.sleep(10)
    scroll_by_func(driver)

    time.sleep(random.random() * 3)
    try:
        content = driver.find_element(By.CSS_SELECTOR, 'blockquote p a')
        new_url = content.get_property('href')
    except:
        print(f'failed to find the new url in {url}')
        new_url = None
    print(new_url, content.text)
    # driver.save_screenshot('isrpa-api-doc.png')
    # article-content
    driver.quit()
    return new_url
```

```python
new_url = get_real_url(url)
if new_url:
    get_article(new_url)
```

```python
# df.loc[articles, 'url'].apply(lambda x: print(x))
```

```python
df.loc[articles, 'content'] = df.loc[articles, 'url'].apply(lambda x: get_article(get_real_url(x)))
```

```python
df.to_json('isrpa-api-v2.json')
```
