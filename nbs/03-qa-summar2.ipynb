你是一个专业的程序员，请根据以下的函数功能描述提取以下信息。
1，为该函数取一个英文的函数名字；
2，提取函数的参数列表；
3，为该函数生成简短描述，包含函数的主要功能，以及什么情况下调用这个函数；
4，提供一个或者两个调用示例。
最终输出为json格式，包含name. parameters, description, usages四项内容。

```python
import os
import openai
from dotenv import load_dotenv
import guidance
# Set the default llm. Could also pass a different one as argument to guidance(), with guidance(llm=...)


load_dotenv()

openai.api_type = os.getenv("OPENAI_API_TYPE")
openai.api_key = os.getenv("OPENAI_API_KEY")
openai.api_base = os.getenv("OPENAI_DEPLOYMENT_ENDPOINT")
openai.api_version = os.getenv("OPENAI_DEPLOYMENT_VERSION")

engine_name = os.getenv("OPENAI_DEPLOYMENT_NAME")
guidance.llm = guidance.llms.OpenAI('gpt-3.5-turbo', 
                                    api_type = os.getenv("OPENAI_API_TYPE"),
                                    api_key = os.getenv("OPENAI_API_KEY"),
                                    api_base = os.getenv("OPENAI_DEPLOYMENT_ENDPOINT"),
                                    api_version = os.getenv("OPENAI_DEPLOYMENT_VERSION"),
                                    deployment_id = engine_name,
                                    temperature=0.1)
```

```python
prompt = guidance(
'''{{#system~}}
You are a helpful assistant.
{{~/system}}
{{#user~}}
{{conversation_question}}
{{~/user}}
{{#assistant~}}
{{gen 'response'}}
{{~/assistant}}''')
r = prompt(conversation_question='What is the meaning of life?')
r
```
